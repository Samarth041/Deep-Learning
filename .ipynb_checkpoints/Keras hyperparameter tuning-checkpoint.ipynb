<h3>how to select appropriate optimsier</h3>

```python
pip install -U keras-tuner
```

```python
import keras_tuner as kt
```

```python
#first we have to make function then tuner object and have to pass that function in that object
```

```python
def build_model(hp):    #hp:-hyperparameter - it is class object which we are getting automatically
    model=Sequential()
    model.add(Input(shape=(8,))) ## input layer (8 features)
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(
        optimizer=hp.Choice('optimizer', values=['adam', 'sgd', 'rmsprop', 'adadelta']),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

# Keras Tuner will create multiple models,
# each time selecting a different optimizer from the given list
```

```python
#now we have to create tuner object
 
tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy', #вҖңChoose model with highest validation accuracyвҖқ
    max_trials=5 # it is class  ; fisrt parameter:-function name 
)
```

```python
tuner.search(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

<h2>No of neuron should be in our model</h2>

```python
#firstly we have chosen 32 as no. of neurons in layer arbitrarily now we will find best no . of parameters
```

```python
def build_model(hp):
    model = Sequential()

    units = hp.Int(
        'units',      # hyperparameter name
        min_value=8,
        max_value=128,
        step=8 # if u dont want to take big jumps to ye step wala parameter hi hata do
    )

    model.add(Input(shape=(8,)))  # input layer (8 features)
    model.add(Dense(units=units, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model
```

```python
tuner=kt.RandomSearch(build_model,objective='val_accuracy',max_trials=5,directory='mydir')
```

```python
tuner.search(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

```python
best_hp = tuner.get_best_hyperparameters()[0]

print(best_hp.values)
```

```python
best_model = tuner.get_best_models(num_models=1)[0]

print(best_model)
```

```python
best_model.fit(X_train,y_train,batch_size=32,epochs=100,initial_epoch=6) #we are keeping initial epoch 6 bcz we already have run 5 epochs already and we want to maintain data of that so we are starting from 6
```

<h2>How to select no. of layers</h2>

```python
def build_model(hp):
    model = Sequential()

    model.add(Input(shape=(8,)))

    # Tune number of hidden layers: 1 to 5
    num_layers = hp.Int(
        'num_layers',
        min_value=1,
        max_value=5
    )

    for i in range(num_layers):
        model.add(
            Dense(
                units=32,
                activation='relu'
            )
        )

    model.add(Dense(1, activation='sigmoid'))

    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model
```

```python
tuner=kt.RandomSearch(build_model,objective='val_accuracy',max_trials=3,directory='mydir',project_name='num_layers')
```

```python
tuner.search(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model=tuner.get_best_models(num_models=1)[0]
```

```python
model.fit(X_train,y_train,epochs=100,initial_epoch=6,validation_data=(X_test,y_test))
```

<h3>Complex</h3>

```python
def build_model(hp):

    model = Sequential()

    counter = 0

    num_layers = hp.Int('num_layers', min_value=1, max_value=10)

    for i in range(num_layers):

        if counter == 0:

            model.add(
                Dense(
                    units=hp.Int(
                        'units_' + str(i),
                        min_value=8,
                        max_value=128,
                        step=8
                    ),
                    activation=hp.Choice(
                        'activation_' + str(i),
                        values=['relu', 'tanh', 'sigmoid']
                    ),
                    input_dim=8
                )
            )

        else:

            model.add(
                Dense(
                    units=hp.Int(
                        'units_' + str(i),
                        min_value=8,
                        max_value=128,
                        step=8
                    ),
                    activation=hp.Choice(
                        'activation_' + str(i),
                        values=['relu', 'tanh', 'sigmoid']
                    )
                )
            )

        counter += 1

    model.add(Dense(1, activation='sigmoid'))

    model.compile(
        optimizer=hp.Choice(
            'optimizer',
            values=['adam', 'rmsprop', 'sgd', 'nadam', 'adadelta']
        ),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model
```

```python
tuner=kt.RandomSearch(build_model,
                      objective='val_accuracy',
                      max_trials=3,
                      directory='mydir',
                      project_name='final'
                        
                     )
```

```python
tuner.search(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model=tuner.get_best_models(num_models=1)[0]
```

```python
model.fit(X_train,y_train,epochs=200,initial_epoch=6,validation_data=(X_test,y_test))
```

```python
from keras import Sequential
from keras.layers import Dense, Dropout, Input

def build_model(hp):

    model = Sequential()

    counter = 0

    num_layers = hp.Int('num_layers', min_value=1, max_value=10)

    for i in range(num_layers):

        units = hp.Int(
            'units_' + str(i), #units:-no of neurons in dense layer
            min_value=8,
            max_value=128,
            step=8
        )

        activation = hp.Choice(
            'activation_' + str(i),
            values=['relu', 'tanh', 'sigmoid']
        )

        dropout_rate = hp.Float(
            'dropout_' + str(i),
            min_value=0.0,
            max_value=0.5,
            step=0.1
        )

        if counter == 0:
            model.add(Dense(units=units,
                            activation=activation,
                            input_dim=8))
        else:
            model.add(Dense(units=units,
                            activation=activation))

        #  Dropout added after every hidden layer
        model.add(Dropout(rate=dropout_rate))

        counter += 1

    model.add(Dense(1, activation='sigmoid'))

    model.compile(
        optimizer=hp.Choice(
            'optimizer',
            values=['adam', 'rmsprop', 'sgd', 'nadam', 'adadelta']
        ),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model
```
